```python
# read the data

heart_health = pd.read_csv('../heart_health_smoker/data/Heart_health.csv', index_col = 0)
heart_health.head()

```

```python
# split the data
train_df, test_df = train_test_split(heart_health, test_size=0.2, random_state=123)
train_df.head()
```

```python
heart_summary = train_df.describe(include="all")
heart_summary
```

```python
X_train = train_df[['Cholesterol(mg/dL)','Glucose(mg/dL)']]
y_train = train_df['Smoker']
X_test = test_df[['Cholesterol(mg/dL)','Glucose(mg/dL)']]
y_test = test_df['Smoker']

X_test.head()
```

```python
# Decision Tree Classifier
X_binary = X_train.copy()
columns = ['Cholesterol(mg/dL)','Glucose(mg/dL)']
for col in columns:
    X_binary[col] = X_binary[col].apply(lambda x: 1 if x >= 215 else 0)

X_binary.head()
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier() # Create a decision tree
model.fit(X_binary, y_train) # Fit a decision tree
model.score(X_binary, y_train) # Assess the model
```

```python
from sklearn.linear_model import LinearRegression

weight_train = train_df['Weight(kg)'].values.reshape(-1, 1)  # Reshape for sklearn
height_train = train_df['Height(cm)']

weight_test = test_df['Weight(kg)'].values.reshape(-1, 1)  # Reshape for sklearn
height_test = test_df['Height(cm)']

# Initialize the Linear Regression model
lin_reg = LinearRegression()

# Fit the model to your data
lin_reg.fit(weight_train, height_train)

# Assuming X_test is your test dataset
height_pred = lin_reg.predict(weight_test)
height_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# Calculate MSE and R-squared
mse = mean_squared_error(height_test, height_pred)
r2 = r2_score(height_test, height_pred)

print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```
